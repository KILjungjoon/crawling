```
import requests as rq
from bs4 import BeautifulSoup as bs
from datetime import datetime as dt
now=dt.now()
import time

# ﻿1) 크롤링 하기

# 타겟 사이트 정의
targetSite='http://stu-c.blog.jp/archives/4785033.html'

# 사이트 연결
step01 = rq.get('http://stu-c.blog.jp/archives/4785033.html')
print(step01)           # <Response [200]>으로 정상. 만약 <Response [406]> => 헤더 정보 때문에 읽어오지 못함

# <Response [406]> 문제 해결 방법 : 헤더 정보를 설정한 후 읽어 와야 한다.
# 여기 내 code로 가기-> https://colab.research.google.com/drive/1P_3wgXqrIVG_pXtvPSGa2h-ZM3LGbvY1#scrollTo=a90WoyRBzxeH
```

```
# 사이트 html 읽기 
step02= step01.text
print(step02)
```

```
# beautifulSoup 활용
step03=bs(step02,'html.parser')
print(step03)
```

```
# 동일 속성을 가진 여러 태그를 가져와야 하므로 find_all을 사용.
# get_text()는 태그를 제외한 실제 얻고자 하는 text만 가져오는 함수.
# class="han han-r"
zhongwen=step03.find_all('div',{'class' :'han han-r'})   
for i in zhongwen :
  print(i.get_text())
```

```
# class="pin pin-r"
pinyin=step03.find_all('div',{'class' :'pin pin-r'})   
for i in pinyin :
  print(i.get_text())
```
